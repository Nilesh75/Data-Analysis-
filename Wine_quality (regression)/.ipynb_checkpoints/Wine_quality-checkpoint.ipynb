### Data Wrangling
Now after the analysis of the data it is useful to maybe reduce the number of categories of target variable quality. Here we divide it in 2 categories namely 'good' and 'bad'.

```python
bins = (2, 6.5, 8)
group_names = ['bad', 'good']
wine['quality'] = pd.cut(wine['quality'], bins = bins, labels = group_names)

label_quality = LabelEncoder()
wine['quality'] = label_quality.fit_transform(wine['quality'])
wine['quality'].value_counts()
```

### Applying Machine Learning Model
Now finally after processing our data we are ready to train it on relevent models. Later I also did GridSearch for algorithm tuning and manually different variations were also tried for better performance.

```python
test_size = 0.25
seed = 7
score='accuracy'
```

```python
X_all = wine.iloc[:,:11]
y_all = wine.iloc[:,11]

X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size = test_size, random_state= seed)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.fit_transform(X_test)
```

```python
#A function for testing various regression models
def models_R(X_train, y_train, score):
    models = []
    result = []
    
    models.append(('LR', LinearRegression()))
    models.append(('LSVR', LinearSVR()))
    models.append(('SVR', SVR()))
    models.append(('RFR', RandomForestRegressor(n_estimators = 30)))
    models.append(('KNR', KNeighborsRegressor()))
    models.append(('GNB', GaussianNB()))
    models.append(('LDA', LinearDiscriminantAnalysis()))
    
    for name, model in models:
        kfold = model_selection.KFold(n_splits = 10, random_state=seed)
        cv_score = model_selection.cross_val_score(model, X_train_s, y_train, cv=kfold, scoring=score)
        
        result.append((name, cv_score.mean(), cv_score.std()))
    return result
    
```

```python
#A function for testing various Classification models
def models_C(X_train, y_train, score):
    models = []
    result = []
    
    models.append(('RFC', RandomForestClassifier(n_estimators = 100)))
    models.append(('GBC', GradientBoostingClassifier(n_estimators = 100)))
    models.append(('SVM', SVC()))
    models.append(('SGD', SGDClassifier()))
    models.append(('LR', LogisticRegression()))
    models.append(('KNC', KNeighborsClassifier()))
    models.append(('GNB', GaussianNB()))
    models.append(('LDA', LinearDiscriminantAnalysis()))
        
    for name, model in models:
        kfold = model_selection.KFold(n_splits = 10, random_state=seed)
        cv_score = model_selection.cross_val_score(model, X_train_s, y_train, cv=kfold, scoring=score)
        
        result.append((name, cv_score.mean(), cv_score.std()))
    return result
```
